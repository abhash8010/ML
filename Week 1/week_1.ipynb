```python
import numpy as np
import pandas as pd
```

```python
x = np.array([2,4]) 
y = np.array([1,5])

# Vector Addition
print(x+y)

# Scalar Product
print(3 * x)

# Dot Product
print(np.dot(x,y))

# Magnitude of a vector
print(np.linalg.norm(x))

# Cosine Similarity
# Dot product over magnitude of vector a and vector b
cosine_similarity = np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y))
print(f'{cosine_similarity:.2f}')
```

```python
X = np.array([
    [1,2,3],
    [5,2,7]
]
)
Y = np.array([
    [1],
    [7],
    [9]
]
)
XY = np.dot(X,Y)
print(XY)
print(f'Shape of X: {X.shape}, Shape of Y: {Y.shape}, and Shape of XY: {XY.shape}')
```

```python
X = np.array([
    [85, 90, 78, 92],
    [70, 65, 80, 75],
    [95, 88, 92, 97]
])

w = np.array([0.2, 0.3, 0.1, 0.4])
prediction = X @ w
print(prediction)
print(prediction.shape)
```

```python
# Manual matrix multiplication
X = np.random.randn(3,4)
Y = np.random.randn(4,2)

result = np.zeros((X.shape[0], Y.shape[1]))

def multiplication(mat_a, mat_b):
    for i in range(mat_a.shape[0]):
        for j in range(mat_b.shape[1]):
            for k in range(mat_a.shape[1]):
                result[i][j] += mat_a[i][k] * mat_b[k][j]
    return result
output = multiplication(X,Y)
print()
print(output)
print(np.allclose(output, X @ Y))
```
